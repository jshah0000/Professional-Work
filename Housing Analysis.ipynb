
This code converts certain text labels in the housing data to numerical values. For example, wherever the data says “yes,” it becomes “1,” and “no” becomes “2.” Similarly, different furnishing conditions (“furnished,” “unfurnished,” “semi-furnished”) are turned into numbers (“1,” “2,” or “3”). By doing this, the data becomes easier for analytical tools and models to handle, since many statistical and machine learning methods require numerical rather than text-based inputs.

```python
# yes = 1, no = 2
#furnished = 1, unfurnished = 2, semi-furnished = 3

def function(column):
    if column=='yes':
        return 1
    else:
        return 2
    
def function1 (col):
    if col == 'furnished':
        return 1
    elif col == 'unfurnished':
        return 2
    else:
        return 3

housing_data[['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning','prefarea']]=housing_data[['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning','prefarea']].applymap(function)

housing_data['furnishingstatus'] = housing_data['furnishingstatus'].apply(function1)
```

```python
housing_data.isna().mean()
```

This step calculates how much each house costs per unit of area (for example, price per square foot or square meter) and saves it in a new column called **`price/area`**. This helps compare properties more fairly by taking into account differences in their size, so you can see if a property’s total price is high or low relative to the space it offers.

```python
#feature engineering
housing_data['price/area'] = housing_data['price']/housing_data['area']
housing_data
```

This code trains a **simple linear regression model** that uses only the **area** of a house to predict its **price**. Here’s what happens step by step:

1. **Feature Selection and Scaling**  
   - The code takes the **`area`** column from the housing data and converts it into a numerical form suitable for modeling, then **scales** it so that large and small area values are placed on a similar scale.

2. **Splitting Data**  
   - The data is divided into two sets: a **training set** (used to teach the model) and a **testing set** (used to check how well the model performs on unseen data).

3. **Building and Training the Model**  
   - A **linear regression** model is created and trained (i.e., it “learns” how area relates to price) using the training set.

4. **Making Predictions & Evaluation**  
   - The trained model **predicts** prices for the test set, and its accuracy is measured:
     - **Mean Squared Error (MSE)**: Shows how far off the predictions are from the actual prices (lower is better).  
     - **R² Score**: Indicates what fraction of price variation is explained by the model (closer to 1 is better).

5. **Results**  
   - The printed **mean squared error** and **R² score** tell us how effectively “area alone” predicts housing prices.  
   - The **model coefficients** show how much price is expected to change based on changes in area.

Essentially, this exercise demonstrates how a basic statistical model (linear regression) can be built using just one property feature—**area**—to make a rough estimate of its **price**.

```python
#linear regression model

import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler

#The dependent variable is area is this case
#step 1 Select all relevant features

scaler = MinMaxScaler()

X = pd.get_dummies(housing_data[['area']])
X = scaler.fit_transform(X)

Y = housing_data[['price']]

#step 2 split data into testing and training sets

X_test, X_train, Y_test, Y_train = train_test_split(X, Y, test_size = 0.2, random_state = 42)

#step 3 Fitting the model

lrm = LinearRegression()

lrm.fit(X_train, Y_train)

#step 4 making predictions evaluating the model

y_pred = lrm.predict(X_test)

mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print('mean squared error:', mse)
print('r2 score:', r2)

coefficients = lrm.coef_

print(coefficients)
```


1. **Mean Squared Error (MSE = 2.51×10^11)**  
   - This figure shows, on average, how far off the model’s predicted prices are from the actual prices. Since the number is quite high, it means that **area alone** doesn’t do a good job at predicting house prices—there are likely many other factors (like location, number of bedrooms, age of the property, etc.) that are not accounted for.

2. **R² Score (≈ 0.185)**  
   - R² describes how much of the variation in house prices the model explains. A score of **0.185** means that around **18.5%** of the price differences can be attributed to the property’s area, leaving more than **80%** unexplained by this single factor. In other words, “area” is not enough on its own to accurately predict housing prices.

3. **Coefficient = ~9,224,054.56**  
   - This value indicates how much the predicted price changes when the area increases by one unit (typically a square foot or square meter). In this data, it suggests a big jump in price per additional unit of area. However, it’s important to note that if other significant features are missing (like neighborhood or number of bedrooms), the model may inflate this relationship.

### Key Takeaways
- With only **one feature—area**—our predictive power is limited.
- The **high MSE** and **low R²** highlight that price is influenced by many more factors.
- Including extra features (location, amenities, etc.) would generally yield a more accurate and meaningful model.

This code demonstrates how a **linear regression model** can predict **house prices** based on four features:
1. **Number of Bedrooms**  
2. **Number of Bathrooms**  
3. **Presence of a Guest Room**  
4. **Furnishing Status** (e.g., furnished, semi-furnished)

Here’s the process:

1. **Feature Selection**  
   - The model uses the four selected features to learn how they impact price.

2. **Splitting the Data**  
   - The dataset is divided into two parts:  
     - **Training Set** – used to teach the model.  
     - **Test Set** – used to evaluate how well the model works on new data.

3. **Building and Training the Model**  
   - A linear regression model is created and “fits” itself to the training data, learning how each feature influences price.

4. **Making Predictions & Evaluating Performance**  
   - The trained model predicts house prices for the test set.  
   - **Mean Squared Error (MSE)** and **R² Score** are calculated:  
     - **MSE** indicates how far off predictions are, on average (lower is better).  
     - **R²** shows how much of the price variation is explained by these features (closer to 1 is better).  
   - The final printout shows the **model coefficients**, revealing how strongly each feature (bedrooms, bathrooms, guest room, furnishing status) affects the predicted house price.

```python
#In this instant the dependent variables are 'bedrooms','bathrooms','guestroom','furnishingstatus'
#step 1 Select all relevant features

X = housing_data[['bedrooms','bathrooms','guestroom','furnishingstatus']]

Y = housing_data[['price']]

#step 2 split data into testing and training sets

X_test, X_train, Y_test, Y_train = train_test_split(X, Y, test_size = 0.2, random_state = 42)

#step 3 Fitting the model

lrm = LinearRegression()

lrm.fit(X_train, Y_train)

#step 4 making predictions evaluating the model

y_pred = lrm.predict(X_test)

mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)

print('mean squared error:', mse)
print('r2 score:', r2)

coefficients = lrm.coef_
print('bedrooms:', coefficients[0][0],
     'bathrooms:', coefficients[0][1],
      'guestroom:', coefficients[0][2],
      'furnishingstatus:', coefficients[0][3])

```


1. **Mean Squared Error (MSE ≈ 2.07×10^11)**  
   - This number tells us how far off, on average, the predicted prices are from the real prices. While it’s lower than some of our previous single-feature attempts, it remains quite large overall. This suggests our model still leaves room for improvement, and more features (like location or total area) may help reduce this error.

2. **R² Score (≈ 0.328)**  
   - An R² of roughly **0.328** means these four features together explain about **32.8%** of the variation in house prices. It’s higher than when we used only one feature (area), but it still indicates that the majority of price differences are caused by factors outside these four variables.

3. **Coefficients**  
   - **Bedrooms: ~757,510**  
     - Each additional bedroom adds roughly 757,510 units to the predicted price, all else being equal.  
   - **Bathrooms: ~1,397,789**  
     - Each additional bathroom increases the predicted price by about 1.4 million units, making it the strongest positive predictor among these features.  
   - **Guestroom: ~-784,163**  
     - Surprisingly, having a “guestroom” is associated with a negative coefficient here. This can happen if the feature interacts with others or if there are data quirks (e.g., most “guestroom” homes in the dataset might be older or in less expensive areas).  
   - **Furnishing Status: ~60,820**  
     - A higher furnishing status (e.g., furnished vs. unfurnished) adds about 60,820 to the predicted price. This smaller value indicates furnishing has less impact on price than bedrooms or bathrooms in this dataset.

### Key Takeaways
- **Bathrooms** contribute the largest price jump among these four features.
- **Guestroom** shows a negative coefficient, which might point to specific patterns in the data worth exploring further.
- Overall, these variables explain around **32.8%** of house price differences, so adding more relevant features (location, size, property age, etc.) may lead to better predictions.
